```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
#load
data=pd.read_csv('customer_data_records.csv')
```

```python
data.columns
```

```python
x=data[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
#clustering
kmeans=KMeans(n_clusters=2,random_state=0)
kmeans.fit(x)
```

```python
data['cluster']=kmeans.labels_
print(data)
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=data,y='Annual Income (k$)',x='Spending Score (1-100)',hue='cluster')
plt.title('Income vs spending')
plt.ylabel('Annual Income (k$)')
plt.xlabel('Spending Score (1-100)')
plt.show()
```

```python
data.columns
```

```python
#how customers earn
y=data[['CustomerID','Annual Income (k$)']]
kmeans.fit(y)
data['cluster']=kmeans.labels_
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=data,y='Annual Income (k$)',x='CustomerID',hue='cluster')
plt.title('customer income')
plt.ylabel('Annual Income (k$)')
plt.xlabel('CustomerID')
plt.show()
```

```python
#how customers spend
a=data[['CustomerID','Spending Score (1-100)']]
kmeans.fit(a)
data['cluster']=kmeans.labels_
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=data,y='Spending Score (1-100)',x='CustomerID',hue='cluster')
plt.title('Customer spending')
plt.ylabel('Spending Score (1-100)')
plt.xlabel('CustomerID')
plt.show()
```
